```python
# Example of a Fully Connected Neural Network
# We will implement a neural network with:
# 2 input neurons
# 3 neurons in the hidden layer
# 1 output neuron

import numpy as np

# Inputs: 2 values
inputs = [1.0, 2.0]

# Weights for the hidden layer (3 neurons, 2 inputs each)
weights_hidden = [
    [0.2, 0.8],  # Weights for the first neuron in the hidden layer
    [-0.5, 1.0],  # Weights for the second neuron in the hidden layer
    [0.7, -0.3]   # Weights for the third neuron in the hidden layer
]

# Biases for the hidden layer (one for each neuron)
biases_hidden = [2.0, 3.0, 0.5]

# Calculate outputs for the hidden layer
hidden_layer_outputs = [
    np.dot(inputs, weights_hidden[i]) + biases_hidden[i] for i in range(len(weights_hidden))
]

# Convert np.float64 values to regular Python floats
hidden_layer_outputs = [float(output) for output in hidden_layer_outputs]

# Weights for the output layer (1 neuron, 3 inputs from the hidden layer)
weights_output = [0.5, -1.5, 1.0]  # Weights for the output neuron
bias_output = 1.0

# Calculate the final output
output = np.dot(hidden_layer_outputs, weights_output) + bias_output

print(hidden_layer_outputs, output)
```

5. Using loop option to get the weights, inputs and biases then feed into the neurons.

```python
# Using loops to calculate the output for the fully connected neural network

# Randomly initialize weights and biases for the hidden layer
inputs = [1.0, 2.0, 3.0, 2.5]
weights_hidden = [[random.uniform(-1, 1) for _ in range(4)] for _ in range(3)]
biases_hidden = [random.uniform(-2, 2) for _ in range(3)]

# Compute outputs for the hidden layer using loops
hidden_layer_outputs = []
for i in range(len(weights_hidden)):
    neuron_output = 0
    for j in range(len(inputs)):
        neuron_output += inputs[j] * weights_hidden[i][j]
    neuron_output += biases_hidden[i]  # Add bias
    hidden_layer_outputs.append(neuron_output)

# Randomly initialize weights and bias for the output layer
weights_output = [random.uniform(-1, 1) for _ in range(3)]
bias_output = random.uniform(-2, 2)

# Compute the final output using loops
output = 0
for i in range(len(hidden_layer_outputs)):
    output += hidden_layer_outputs[i] * weights_output[i]
output += bias_output  # Add bias

weights_hidden, biases_hidden, weights_output, bias_output, hidden_layer_outputs, output
```

Activation functions.

```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example usage with an array of values
x = np.array([0.5, -1.0, 2.0])
output = sigmoid(x)

output
```

```python
import numpy as np

# ReLU activation function
def relu(x):
    return np.maximum(0, x)

# Example usage with an array of values
x = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])
output = relu(x)

output
```

```python
# Step function definition
def step_function(x, threshold=0):
    if x <= threshold:
        return 0
    else:
        return 1

# Example usage with a value
x = 0.5
output = step_function(x)

output
```

```python
# Step activation function
def step_function(x, threshold=0):
    return np.array([1 if value > threshold else 0 for value in x])

# Example usage
x = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])
output = step_function(x)

output
```

Solve gate problem using neural network.

```python
# 2-bit AND Gate Example
import numpy as np

# Define the inputs and corresponding labels for the AND gate
inputs = np.array([[0, 0],
                   [0, 1],
                   [1, 0],
                   [1, 1]])
labels = np.array([0, 0, 0, 1])

# Define the weights and biases for the neural network
weights = np.array([0.5, 0.5])
bias = -0.7

# Define the activation function (step function)
def step_function(x):
    return 0 if x <= 0 else 1

# Perform the forward pass to compute the outputs
outputs = []
for input_row in inputs:
    weighted_sum = np.dot(input_row, weights) + bias
    output = step_function(weighted_sum)
    outputs.append(output)

# Compare the predicted outputs with the actual labels
correct_predictions = np.sum(outputs == labels)

# Print the results
inputs, labels, outputs, correct_predictions
```

```python
# 2-bit Randomly initialize weights and biases for the AND gate
import random

# Randomly adjust weights and bias for the AND gate neural network
weights = np.array([random.uniform(-1, 1) for _ in range(2)])
bias = random.uniform(-1, 1)

# Perform the forward pass to compute the outputs with the new weights and bias
outputs = []
for input_row in inputs:
    weighted_sum = np.dot(input_row, weights) + bias
    output = step_function(weighted_sum)
    outputs.append(output)

# Compare the predicted outputs with the actual labels
correct_predictions = np.sum(outputs == labels)

weights, bias, outputs, correct_predictions
```

```python
import numpy as np

# Define the sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    fx = sigmoid(x)
    return fx * (1 - fx)

# Mean Squared Error loss function
def mean_squared_error_loss(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

class SimpleNeuralNetwork:
    '''
    A simple neural network with:
      - 2 inputs
      - a hidden layer with 2 neurons (hidden_neuron_1, hidden_neuron_2)
      - an output layer with 1 neuron (output_neuron)
    '''
    def __init__(self):
        # Initialize weights randomly
        self.weight_1 = np.random.normal()
        self.weight_2 = np.random.normal()
        self.weight_3 = np.random.normal()
        self.weight_4 = np.random.normal()
        self.weight_5 = np.random.normal()
        self.weight_6 = np.random.normal()

        # Initialize biases randomly
        self.bias_1 = np.random.normal()
        self.bias_2 = np.random.normal()
        self.bias_output = np.random.normal()

    def feedforward(self, inputs):
        # Perform a feedforward computation
        hidden_neuron_1 = sigmoid(self.weight_1 * inputs[0] + self.weight_2 * inputs[1] + self.bias_1)
        hidden_neuron_2 = sigmoid(self.weight_3 * inputs[0] + self.weight_4 * inputs[1] + self.bias_2)
        output_neuron = sigmoid(self.weight_5 * hidden_neuron_1 + self.weight_6 * hidden_neuron_2 + self.bias_output)
        return output_neuron

    def train(self, data, target_outputs):
        # Training parameters
        learning_rate = 0.1
        epochs = 1000  # Number of times to loop through the entire dataset

        for epoch in range(epochs):
            for inputs, target in zip(data, target_outputs):
                # --- Perform a feedforward pass
                sum_hidden_1 = self.weight_1 * inputs[0] + self.weight_2 * inputs[1] + self.bias_1
                hidden_neuron_1 = sigmoid(sum_hidden_1)

                sum_hidden_2 = self.weight_3 * inputs[0] + self.weight_4 * inputs[1] + self.bias_2
                hidden_neuron_2 = sigmoid(sum_hidden_2)

                sum_output = self.weight_5 * hidden_neuron_1 + self.weight_6 * hidden_neuron_2 + self.bias_output
                output_neuron = sigmoid(sum_output)
                predicted_output = output_neuron

                # --- Calculate partial derivatives
                error = -2 * (target - predicted_output)

                # Output neuron
                d_output_d_w5 = hidden_neuron_1 * sigmoid_derivative(sum_output)
                d_output_d_w6 = hidden_neuron_2 * sigmoid_derivative(sum_output)
                d_output_d_bias_output = sigmoid_derivative(sum_output)

                d_output_d_hidden_1 = self.weight_5 * sigmoid_derivative(sum_output)
                d_output_d_hidden_2 = self.weight_6 * sigmoid_derivative(sum_output)

                # Hidden neuron 1
                d_hidden_1_d_w1 = inputs[0] * sigmoid_derivative(sum_hidden_1)
                d_hidden_1_d_w2 = inputs[1] * sigmoid_derivative(sum_hidden_1)
                d_hidden_1_d_bias_1 = sigmoid_derivative(sum_hidden_1)

                # Hidden neuron 2
                d_hidden_2_d_w3 = inputs[0] * sigmoid_derivative(sum_hidden_2)
                d_hidden_2_d_w4 = inputs[1] * sigmoid_derivative(sum_hidden_2)
                d_hidden_2_d_bias_2 = sigmoid_derivative(sum_hidden_2)

                # --- Update weights and biases using gradient descent
                self.weight_1 -= learning_rate * error * d_output_d_hidden_1 * d_hidden_1_d_w1
                self.weight_2 -= learning_rate * error * d_output_d_hidden_1 * d_hidden_1_d_w2
                self.bias_1 -= learning_rate * error * d_output_d_hidden_1 * d_hidden_1_d_bias_1

                self.weight_3 -= learning_rate * error * d_output_d_hidden_2 * d_hidden_2_d_w3
                self.weight_4 -= learning_rate * error * d_output_d_hidden_2 * d_hidden_2_d_w4
                self.bias_2 -= learning_rate * error * d_output_d_hidden_2 * d_hidden_2_d_bias_2

                self.weight_5 -= learning_rate * error * d_output_d_w5
                self.weight_6 -= learning_rate * error * d_output_d_w6
                self.bias_output -= learning_rate * error * d_output_d_bias_output

            # --- Calculate total loss every 10 epochs
            if epoch % 10 == 0:
                predictions = np.apply_along_axis(self.feedforward, 1, data)
                loss = mean_squared_error_loss(target_outputs, predictions)
                print(f"Epoch {epoch} loss: {loss:.3f}")

# Define the dataset
data = np.array([
    [-2, -1],  # Alice
    [25, 6],   # Bob
    [17, 4],   # Charlie
    [-15, -6], # Diana
])
target_outputs = np.array([
    1,  # Alice
    0,  # Bob
    0,  # Charlie
    1,  # Diana
])

# Train the neural network
network = SimpleNeuralNetwork()
network.train(data, target_outputs)

# Make some predictions
emily = np.array([-7, -3])  # Example: Emily's features
frank = np.array([20, 2])   # Example: Frank's features
print("Emily: %.3f" % network.feedforward(emily))  # Expected output close to 1
print("Frank: %.3f" % network.feedforward(frank))  # Expected output close to 0
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming we have a list of loss values recorded during training
loss_values = []

# Define sigmoid and sigmoid_derivative functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    fx = sigmoid(x)
    return fx * (1 - fx)

# Mean Squared Error loss function
def mean_squared_error_loss(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

class SimpleNeuralNetwork:
    def __init__(self):
        self.weight_1 = np.random.normal()
        self.weight_2 = np.random.normal()
        self.weight_3 = np.random.normal()
        self.weight_4 = np.random.normal()
        self.weight_5 = np.random.normal()
        self.weight_6 = np.random.normal()
        self.bias_1 = np.random.normal()
        self.bias_2 = np.random.normal()
        self.bias_output = np.random.normal()

    def feedforward(self, inputs):
        hidden_neuron_1 = sigmoid(self.weight_1 * inputs[0] + self.weight_2 * inputs[1] + self.bias_1)
        hidden_neuron_2 = sigmoid(self.weight_3 * inputs[0] + self.weight_4 * inputs[1] + self.bias_2)
        output_neuron = sigmoid(self.weight_5 * hidden_neuron_1 + self.weight_6 * hidden_neuron_2 + self.bias_output)
        return output_neuron

    def train(self, data, target_outputs):
        learning_rate = 0.1
        epochs = 1000

        for epoch in range(epochs):
            for inputs, target in zip(data, target_outputs):
                sum_hidden_1 = self.weight_1 * inputs[0] + self.weight_2 * inputs[1] + self.bias_1
                hidden_neuron_1 = sigmoid(sum_hidden_1)

                sum_hidden_2 = self.weight_3 * inputs[0] + self.weight_4 * inputs[1] + self.bias_2
                hidden_neuron_2 = sigmoid(sum_hidden_2)

                sum_output = self.weight_5 * hidden_neuron_1 + self.weight_6 * hidden_neuron_2 + self.bias_output
                output_neuron = sigmoid(sum_output)
                predicted_output = output_neuron

                error = -2 * (target - predicted_output)

                d_output_d_w5 = hidden_neuron_1 * sigmoid_derivative(sum_output)
                d_output_d_w6 = hidden_neuron_2 * sigmoid_derivative(sum_output)
                d_output_d_bias_output = sigmoid_derivative(sum_output)

                d_output_d_hidden_1 = self.weight_5 * sigmoid_derivative(sum_output)
                d_output_d_hidden_2 = self.weight_6 * sigmoid_derivative(sum_output)

                d_hidden_1_d_w1 = inputs[0] * sigmoid_derivative(sum_hidden_1)
                d_hidden_1_d_w2 = inputs[1] * sigmoid_derivative(sum_hidden_1)
                d_hidden_1_d_bias_1 = sigmoid_derivative(sum_hidden_1)

                d_hidden_2_d_w3 = inputs[0] * sigmoid_derivative(sum_hidden_2)
                d_hidden_2_d_w4 = inputs[1] * sigmoid_derivative(sum_hidden_2)
                d_hidden_2_d_bias_2 = sigmoid_derivative(sum_hidden_2)

                self.weight_1 -= learning_rate * error * d_output_d_hidden_1 * d_hidden_1_d_w1
                self.weight_2 -= learning_rate * error * d_output_d_hidden_1 * d_hidden_1_d_w2
                self.bias_1 -= learning_rate * error * d_output_d_hidden_1 * d_hidden_1_d_bias_1

                self.weight_3 -= learning_rate * error * d_output_d_hidden_2 * d_hidden_2_d_w3
                self.weight_4 -= learning_rate * error * d_output_d_hidden_2 * d_hidden_2_d_w4
                self.bias_2 -= learning_rate * error * d_output_d_hidden_2 * d_hidden_2_d_bias_2

                self.weight_5 -= learning_rate * error * d_output_d_w5
                self.weight_6 -= learning_rate * error * d_output_d_w6
                self.bias_output -= learning_rate * error * d_output_d_bias_output

            # Calculate total loss every epoch and record it
            predictions = np.apply_along_axis(self.feedforward, 1, data)
            loss = mean_squared_error_loss(target_outputs, predictions)
            loss_values.append(loss)

# Define the dataset
data = np.array([
    [-2, -1],  # Alice
    [25, 6],   # Bob
    [17, 4],   # Charlie
    [-15, -6], # Diana
])
target_outputs = np.array([
    1,  # Alice
    0,  # Bob
    0,  # Charlie
    1,  # Diana
])

# Train the neural network and record the loss
network = SimpleNeuralNetwork()
network.train(data, target_outputs)

# Plot the loss over epochs
plt.plot(loss_values)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Reduction Over Time")
plt.show()

```
