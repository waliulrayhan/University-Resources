```python
Xval=data.iloc[:,:-1]
```

```python
Xval
```

```python
Yval=data.iloc[:,-1]
```

```python
Yval
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(Xval, Yval, test_size=0.2, random_state=0)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model
```

```python
model.fit(X_train, Y_train)
```

```python
model.coef_
```

```python
#create dataframe (data, index, columns)
coefficient=pd.DataFrame(model.coef_,Xval.columns, columns=['Coef'])
```

```python
coefficient
```

```python
model.intercept_
```

```python
Y_predict=model.predict(X_test)
```

```python
Y_predict
```

```python
Y_test
```

```python
results=pd.DataFrame({'Actual':Y_test,'Predicted':Y_predict})
```

```python
results
```

```python
plt.scatter(Y_test,Y_predict)
```

```python
from sklearn import metrics 
```

```python
metrics.mean_absolute_error(Y_test,Y_predict)
```

```python
metrics.mean_squared_error(Y_test, Y_predict)
```
